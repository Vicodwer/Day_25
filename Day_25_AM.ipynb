# Part- B

Filter Dataset with Multiple Conditions

```python
filtered_df = df[(df['view_count']>100000) & (df['like_count']> 10000)]
filtered_df.head()
```

Create New Column

```python
df['Engagement_Rate'] = (df['like_count'] / df['view_count'])  * 100
df.head()
```

Sort Dataset

```python
df_sorted = df.sort_values(by='view_count', ascending = False)
df_sorted.head()
```

Group Data & Compute Mean

```python
grouped = df.groupby('category_id')['view_count'].mean()
grouped
```

# Part-C

Q1 — What is EDA? Why is it important?

Exploratory Data Analysis (EDA) is the process of analyzing datasets to understand their structure, patterns, distributions, anomalies, and relationships before applying machine learning models.

Why Important?
Detects missing values
Identifies outliers
Understands feature distributions
Guides feature engineering
Prevents garbage-in garbage-out problem

Q2--Coding Question
Filter rows where column value > average

```python
average_views = df['view_count'].mean()

above_avg = df[df['view_count'] > average_views]
above_avg.head()
```

Q3 — What insights can we get from describe()?

describe() gives:

Count → Number of non-null values
Mean → Central tendency
Std → Variability
Min & Max → Range
25%, 50%, 75% → Quartiles
Insights:
Detect skewness (mean vs median difference)
Identify outliers (max far from 75%)
Understand spread of data
Compare feature scales
